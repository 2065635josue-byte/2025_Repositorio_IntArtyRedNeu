Cargando librerías a utilizar para la actividad fundamental

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Se cargan los datos del archivo CSV enviado por el profesor.

```python
basic_dataset = pd.read_csv('/content/ACT5/robot_sensors.csv')
```

Mostar las primeras 10 filas de los datos en el archivo CSV

```python
print(basic_dataset.head(10))
```

Se obtiene la información sobre las columnas y el tipo de datos de cada una de dichas columnas.

```python
print(basic_dataset.info())
```

Detección de valores nulos (missing values) en alguna columna.

```python
print(basic_dataset.isnull().sum())
```

Filtación de valores nulos en la columna de distancia

```python
basic_dataset['Distancia (cm)'] = basic_dataset['Distancia (cm)'].fillna(basic_dataset['Distancia (cm)'].mean())
```

```python
print(basic_dataset.isnull().sum())
```

Filtracion de valoes nulos en la columna velocidad

```python
basic_dataset['Velocidad (cm/s)'] = basic_dataset['Velocidad (cm/s)'].fillna(basic_dataset['Velocidad (cm/s)'].mean())
```

```python
print(basic_dataset.isnull().sum())
```

Filtracion de valores nulos en la columna temperatura

```python
basic_dataset['Temperatura (°C)'] = basic_dataset['Temperatura (°C)'].fillna(basic_dataset['Temperatura (°C)'].mean())
```

```python
print(basic_dataset.isnull().sum())
```

Filtración de valores atipicos para columnas de distancia, velocidad y temperatura

```python
basic_dataset = basic_dataset[basic_dataset['Distancia (cm)'] >= 0]
```

```python
basic_dataset = basic_dataset[basic_dataset['Velocidad (cm/s)'] >= 0]
```

```python
basic_dataset = basic_dataset[basic_dataset['Temperatura (°C)'] >= 0]
```

Normalizacion de valores utilizando min-max scaling

```python
normalized_data2= (basic_dataset - np.min(basic_dataset)) / (np.max(basic_dataset) - np.min(basic_dataset))
```

Generación de gráfica tiempo - temperatura

```python
plt.scatter(normalized_data2['Tiempo (s)'], normalized_data2['Temperatura (°C)'], color='blue', label="Datos Aleatorios")  # Dibuja el gráfico de dispersión
plt.title("Gráfico de Dispersión")  # Título
plt.xlabel("Tiempo")  # Etiqueta del eje x
plt.ylabel("Temperatura")  # Etiqueta del eje y
plt.legend()  # Muestra la leyenda
plt.grid(True)  # Muestra la cuadrícula
plt.show()  # Muestra el gráfico
```

Generación de gráfica tiempo - distancia

```python
plt.scatter(normalized_data2['Tiempo (s)'], normalized_data2['Distancia (cm)'], color='blue', label="Datos Aleatorios")  # Dibuja el gráfico de dispersión
plt.title("Gráfico de Dispersión")  # Título
plt.xlabel("Tiempo")  # Etiqueta del eje x
plt.ylabel("Distancia")  # Etiqueta del eje y
plt.legend()  # Muestra la leyenda
plt.grid(True)  # Muestra la cuadrícula
plt.show()  # Muestra el gráfico
```
